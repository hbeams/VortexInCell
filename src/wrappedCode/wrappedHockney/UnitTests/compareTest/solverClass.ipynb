```python
import numpy as np
import array as arr
import matplotlib.pyplot as plt
from scipy import signal
import Hockney
import copy

class PoissonSolver:
    def __init__(self):
        self.data = []
        
    def __init__(self, a_dx, a_M, Center = None):
        self._m_dx = a_dx;
        self._m_M = a_M;
        
    def solve(self, RHSArray, fullGrid = True):
        
        if (not fullGrid):
            #pad array with zeros to be full
            #This isn't correct but I don't want to deal with this right now
            inputArray = copy.deepcopy(RHSArray);
        else:
            #Not necessary to do a deep copy here but I want to momentarily be transparent about data being overwritten 
            inputArray = copy.deepcopy(RHSArray);
        
        Hockney.convolve(self._m_dx, self._m_M, inputArray)
        
        return inputArray

    
def cosineBell(r):
    power = 8
    if r < r0:
        return np.cos(np.pi/2*r)**power
    else: 
        return 0.0
    
def gaussianCharge(r):
     if r < r0:
        return np.sqrt(2/np.pi)*np.exp(-1/2*(r)**2)
     else: 
        return 0.0
    
def pointCharge(r):
    if r == 0:
        return 1.0
    else: 
        return 0.0
    
```

```python
#Grid Properties
L = 1;
xCen = L/2
yCen = xCen
r0 = L/4
```

```python
M=8
gridSize = 2**M

x_grid = np.linspace(0, L, gridSize+1)
dx_4 = np.abs(x_grid[0] - x_grid[1])
print(dx_4)
RHSArray = np.zeros((gridSize+1, gridSize+1))

for indexX, X in np.ndenumerate(x_grid):
    for indexY, Y in np.ndenumerate(x_grid):
        RHSArray[indexX, indexY] = pointCharge(np.sqrt((X - xCen)**2+(Y- yCen)**2)/r0)        

PS = PoissonSolver(dx_4, M)
soln_4 = PS.solve(RHSArray)


plt.imshow(RHSArray, extent=[0, L, 0, L])
plt.colorbar(orientation='vertical')
plt.show()
plt.imshow(soln_4, extent=[0, L, 0, L])
plt.colorbar(orientation='vertical')
plt.show()
```
